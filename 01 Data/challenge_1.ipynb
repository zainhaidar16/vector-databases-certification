## Draw Out These Vectores
### Given 3D vectors [40,120,60] and [60,50,90]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean, cityblock, cosine
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Define the vectors
vector1 = [40, 120, 60]
vector2 = [60, 50, 90]

ax = plt.figure().add_subplot(projection='3d')

# Draw the vectors
ax.scatter(xs=vector1[0], ys=vector1[1], zs=vector1[2], zdir='z', label='vector1') 
ax.scatter(xs=vector2[0], ys=vector2[1], zs=vector2[2], zdir='z', label='vector2')

# plot origin
ax.scatter(xs=[0], ys=[0], zs=[0], zdir='z', label='origin', c='black')

# Add arrows for better visualization
ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2], color='r')
ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2], color='g')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Set the view
ax.view_init(elev=20., azim=-35, roll=0.)
plt.show()
```

```python
# Calculate Euclidean distance
import numpy as np
euclidean_distance = euclidean(vector1, vector2)

# Calculate Manhattan distance
manhattan_distance = cityblock(vector1, vector2)

# Calculate cosine similarity
cosine_sim = cosine_similarity([vector1], [vector2])[0][0]

# Calculate dot product
dot_product = np.dot(vector1, vector2)

euclidean_distance, manhattan_distance, cosine_sim, dot_product

```

```python
# Come up with a color that has a smaller cosine distance from the first vector
vector3 = [vector1[0]+10, vector1[1]+15, vector1[2] + 5]
```

```python
ax = plt.figure().add_subplot(projection='3d')

# Draw the vectors
ax.scatter(xs=vector1[0], ys=vector1[1], zs=vector1[2], zdir='z', label='vector1') 
ax.scatter(xs=vector2[0], ys=vector2[1], zs=vector2[2], zdir='z', label='vector2')
ax.scatter(xs=vector3[0], ys=vector3[1], zs=vector3[2], zdir='z', label='vector3')  # Add this line

# plot origin
ax.scatter(xs=[0], ys=[0], zs=[0], zdir='z', label='origin', c='black')

# Add arrows for better visualization
ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2], color='r')
ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2], color='g')
ax.quiver(0, 0, 0, vector3[0], vector3[1], vector3[2], color='b')  # Add this line

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Set the view
ax.view_init(elev=20., azim=-35, roll=0.)
plt.show()
```
